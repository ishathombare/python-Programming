```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.arange(1, 12)
y = 3 * x + 5
plt.plot(x, y, label='y = 3x + 5', color='b')

plt.xlabel('x') 
plt.ylabel('y')  

# Add title
plt.title('Plot of y = 3x + 5')

plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.plot(x, y, label='y = sin(x)', color='r')

# Label the axes
plt.xlabel('x')  
plt.ylabel('y') 

# Add title
plt.title('Plot of y = sin(x)')
plt.grid(True)
```

```python
x = np.arange(0, 5 * np.pi, 0.1) 

y_sin = np.sin(x)
y_cos = np.cos(x)
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))  # 2 rows, 1 column
# Plot sin(x) on the first subplot
ax1.plot(x, y_sin, label='y = sin(x)', color='r')
ax1.set_title('Plot of y = sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True)
ax1.legend()
# Plot cos(x) on the second subplot
ax2.plot(x, y_cos, label='y = cos(x)', color='b')
ax2.set_title('Plot of y = cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True)
ax2.legend()
# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
x = np.array([22, 11, 47, 52, 12, 68, 8, 27, 42])
# Create the histogram
plt.hist(x, bins=5, color='blue', edgecolor='black') # Specify number of bins and color

plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
           
plt.savefig('histogram.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

labels = ['first', 'second', 'third', 'forth']
sizes = [215, 130, 245, 210]  # Corrected variable name for consistency
colors = ['gold', 'red', 'green', 'blue']
explode = (0.1, 0, 0, 0) 

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Pie Chart")
plt.savefig('Pie_chart.png')  
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
# Create the box plot
plt.boxplot(data, vert=True, patch_artist=True)  # Set patch_artist to True for filled boxes
# Add title and Labels
plt.title('Box Plot')
plt.xlabel('Standard Deviation')
plt.ylabel("Value")

plt.savefig("boxplot.png")
```
